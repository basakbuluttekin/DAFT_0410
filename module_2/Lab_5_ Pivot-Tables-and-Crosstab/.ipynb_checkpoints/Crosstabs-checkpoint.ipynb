```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

warnings.filterwarnings('ignore')
%matplotlib inline
```

### 1. Import the moovies dataset 

# https://www.kaggle.com/carolzhangdc/imdb-5000-movie-dataset

```python
movies = pd.read_csv("movie_metadata.csv")
```

```python
movies.head()
```

```python
movies.isna().sum()
```

```python
movies = movies.dropna(subset= ['director_name', 'duration', 'director_facebook_likes', 'actor_3_facebook_likes', 'actor_1_facebook_likes', 'genres', 'country', 'actor_2_facebook_likes'])
```

```python
movies.isna().sum()
```

### 2. Use crosstabs to show the number of moovies per country

```python
#pd.crosstab(movies.country, movies.index, aggfunc=len)
numbers = pd.crosstab(movies['country'], movies.index , margins=True, margins_name='Number of Movies')
numbers['Number of Movies'].to_frame().sort_values(by='country', ascending = True)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
directors = pd.crosstab([movies['country'], movies['director_name'], movies['genres']], movies.index , margins=True, margins_name='Number of Movies')
```

```python
directors['Number of Movies'].to_frame().sort_values(by='country', ascending = True)
```

### 4. Average duration of filmes grouped by color

```python
duration = pd.crosstab(movies['color'],movies.index, values = movies['duration'], aggfunc=np.mean, margins=True, margins_name='Average Duration')
```

```python
duration['Average Duration'].to_frame()
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
likes = pd.crosstab([movies['director_name']], movies.index, values = movies['actor_3_facebook_likes']+ movies['actor_2_facebook_likes']+ movies['actor_1_facebook_likes'], aggfunc='mean', margins=True, margins_name='Number of Likes')
```

```python
likes['Number of Likes'].to_frame()
```
