#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10,10))

ax1.set_title("X - Y Relationship")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.plot(x,y)

ax2.set(title="X - Z Relationship", xlabel="X", ylabel="Z")
ax2.plot(x,z)

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10,10))

font1 = {'family':'serif','color':'blue','size':12}
font2 = {'family':'serif','color':'darkred','size':12}
font3 = {'family':'serif','color':'green','size':12}

ax1.plot(x,y)
ax1.set_title('Numbers')
ax1.set_xlabel("X", fontdict = font1)
ax1.set_ylabel("Y", fontdict = font2)
ax1.legend()


ax2.plot(x,z)
ax2.set_title('Numbers')
ax2.set_xlabel("X", fontdict = font1)
ax2.set_ylabel("Z", fontdict = font3)
ax2.legend('Z')


plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
import math
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
y1 = x**2
ax1.set(title = "Normal Scale", xlabel="X", ylabel="X^2")
ax1.plot(x,y1)

y2 = np.exp(x)
ax2.set(title= "Logarithmic", xlabel="X", ylabel="e^x")
ax2.set_yscale('symlog', linthresh=0.1)
ax2.plot(x,y2)


plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
y1 = x**2
ax1.set(title = "Normal Scale", xlabel="X", ylabel="X^2")
ax1.plot(x,y1, label = 'x^2')
ax1.legend()

y2 = np.exp(x)
ax2.set(title= "Logarithmic", xlabel="X", ylabel="e^x")
ax2.set_yscale('symlog', linthresh=0.1)
ax2.plot(x,y2, label = 'e^x')
ax2.legend()

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,4))


ax1.set(title="Monthly Average Steps", xlabel="Months", ylabel="Avg_Steps")
ax1.plot(fitbit.groupby('Months_encoded')['Steps'].mean())



ax2.set(title="Monthly Median Steps", xlabel="Months", ylabel="Median_Steps")
ax2.plot(fitbit.groupby('Months_encoded')['Steps'].median())


ax3.set(title = "Active Days", ylabel="Sum of Steps")
ax3.xaxis.label.set_visible(False)
fitbit.groupby('Work_or_Weekend')['Steps'].sum().plot(kind='bar')
ax3.set_xticklabels(['Weekend', 'Workdays'])


plt.show()
```

## Write a loop to plot 3 scatter plots of the following features:

#### - Minutes Lightly Active vs Steps
#### - Minutes Very Active vs Steps
#### - Minutes Sedentary vs Steps

```python
features = {
    'Minutes Lightly Active': 'Green',
    'Minutes Very Active': 'Orange',
    'Minutes Sedentary': 'Red'
}

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, (feature, color) in enumerate(features.items()):
    ax[i].set(title=feature, xlabel="Minutes", ylabel="Steps")
    ax[i].scatter(fitbit[feature], fitbit['Steps'], c=color)

plt.show()
```
