## Challenge_3

```python
url = 'https://api.github.com/search/code?q=.scavengerhunt'
```

```python
headers = {"Authorization": "basakbuluttekin <  >"}
```

```python
response1 = requests.get('https://api.github.com/search/code?q=.scavengerhunt', headers = headers);
search = response1.json()
search
```

```python
repo_files_names = [item["0001.scavengerhunt"] for item in search["scavengerhunt"]]
repo_files_names.sort()
repo_files_names
repo_files_urls = {item["0001.scavengerhunt"]: item["url"] for item in search["scavengerhunt"]}
```

```python
repo_files = [requests.get(repo_files_urls[scavenger]) for scavenger in repo_files_names]
```

```python
repo_files
```

```python
repo_files_content = [file.json()["content"] for file in repo_files]
repo_files_content
```

```python
import base64
message = [base64.b64decode(b64message).decode('ascii').replace("\n","") for b64message in repo_files_content]
```

```python
" ".join(message)
```
