#### Question 1: Find the questions and answers of last month.

```python
import pandas as pd
from stackapi import StackAPI

SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1677628800, todate=1680307199,)
questions
```

```python
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', fromdate=1677628800, todate=1680307199,)
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1681862400, todate=1681948799, tagged='python', min = 5, sort='votes')
questions
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers',ids=[6784, 6473])
answers
```
